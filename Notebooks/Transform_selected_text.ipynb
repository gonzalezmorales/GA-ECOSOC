```python
import pandas as pd
from dotenv import load_dotenv
import os

# Load the .env file
load_dotenv()


key = os.getenv('KEY')
```

```python

from openai import OpenAI

client = OpenAI(
  api_key=key,  # this is also the default, it can be omitted
)
```

```python
file_path = '../input/selected_paragraphs_concat.xlsx'
```

```python
data = pd.read_excel(file_path, usecols=[0], header=None)[0].tolist()
data[0]
```

```python
text=f"Transform this text to past tense: {data[0]}"
text
```

```python

def to_past_tense(text):
    completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "user", "content": text}
      ]
    )

    return completion.choices[0].message.content
```

```python
to_past_tense(text)
```

```python
processed_text = []

for i, t in enumerate(data):
   
    text=f"Transform this text to past tense: {data[i]}"
    processed_text.append(to_past_tense(text))
    print(f"finished {i+1} of {len(data)}")
```

```python
def save_processed_text(index, text, filename):
    with open(filename, 'a', encoding='utf-8') as file:
        file.write(f"{index}: {text}\n")

def get_last_processed_index(filename):
    encodings = ['utf-8', 'cp1252', 'ISO-8859-1']  # Add other encodings as needed
    for encoding in encodings:
        try:
            with open(filename, 'r', encoding=encoding) as file:
                lines = file.readlines()
                if lines:
                    last_line = lines[-1]
                    last_index = int(last_line.split(':')[0])
                    return last_index
                else:
                    return -1
        except UnicodeDecodeError:
            continue
    return -1  # Return -1 if all encodings fail


output_file = '../output/selected_paragraphs_processed.txt'

last_index_processed = get_last_processed_index(output_file)

for i, t in enumerate(data):
    if i <= last_index_processed:
        continue

    text = f"Transform this text to past tense: {t}"
    processed_text = to_past_tense(text)
    save_processed_text(i, processed_text, output_file)
    print(f"finished {i + 1} of {len(data)}")
```
