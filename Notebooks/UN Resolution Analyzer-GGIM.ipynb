# Import required libraries

```python
from tika import parser
import re
import json
import csv
```

## Define keywords

```python
keywords = ["sustainable development", "2030 Agenda", "data", "indicator", "names", "location", "addresses", "geo"]
```

# Collect list of resolution documents to be analyzed

```python
# https://code-maven.com/listing-a-directory-using-python

import os,sys


output_path = r'C:\Users\L.GonzalezMorales\Documents\GitHub\GA-ECOSOC\ResolutionsGGIM'

#path = r'C:\Users\L.GonzalezMorales\Documents\GitHub\GA-ECOSOC\ResolutionDocs\ECOSOC2018'
#UNBody = 'The Economic and Social Council'
#documentCodePrfx = 'E/RES/2018/'
#session = '2018 session'
#output_json = 'ECOSOC2018.json'
#output_txt = 'ECOSOC2018.txt'

path = r'C:\Users\L.GonzalezMorales\Documents\GitHub\GA-ECOSOC\ResolutionsGGIM'
UNBody = 'The General Assembly'
documentCodePrfx = 'A/RES/'
session = ''
output_json = 'ggim.json'
output_txt = 'ggim.txt'


documents = []
output = []

files = os.listdir(path)

for f in files:
    documents.append(f)
    print(f)
    parsed = parser.from_file(path + '\\' + f)
    raw = parsed['content']
    
    #Remove extra line breaks, unnecessary white space, and other characters
    raw = re.sub('\n\s+','\n',raw).replace(u'\xa0', u' ').replace(u'\u2212',u'-').replace(u'\u200e','').replace(u'\u2011','').replace(u'\ufffd','')
    print(raw)  
     
    #Create an array with each paragraph as an element
    contentElements = raw.split(UNBody,1)[1].splitlines()
    
    metadata =	{
      "resolution": documentCodePrfx + re.search(re.escape(documentCodePrfx) + r'(.*?)\n', raw).group(1),
      "date": re.search('Distr.: General\n(.*?)\n', raw).group(1),
      "title": re.search(r'\n(.+?)\n' + re.escape(UNBody), raw).group(1),
      "content" : contentElements
    }
    
    print(metadata)

    output.append(metadata)
   
    
print("There are " + str(len(documents)) + " documents in the folder.")
```

## Add resolution information to a json file:

```python
#with open('resolutions.json', 'w') as outfile:
with open(output_path + '\\' + output_json, 'w') as outfile:
    json.dump(output,outfile, indent=4 )    

resolutions_json_data = json.load(open(output_path + '\\' + output_json))

for r in resolutions_json_data:
    print(r["resolution"])
    print(r["date"])
    print(r["title"])
    print(r["content"][0])

#Add resolution information to a tab-delimited text file:
outF = open(output_path + '\\' + output_txt, "w")
for r in resolutions_json_data:
    for c in r["content"]:
        has_keywords = ""
        for k in keywords:
            if k.lower() in c.lower():
                has_keywords = has_keywords + k + ", "
        line = r["resolution"]+"\t" \
               + r["date"]+"\t" \
               + r["title"].replace('\t','   ')+"\t" \
               + c.replace('\t','   ')+"\t" \
               + has_keywords
        outF.write(line)
        outF.write("\n")
outF.close()    
```
